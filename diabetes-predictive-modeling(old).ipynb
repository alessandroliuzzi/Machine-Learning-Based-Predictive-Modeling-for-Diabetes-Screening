```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np



#imports for PIPELINE
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

#MLP - MODEL 1
from sklearn.neural_network import MLPClassifier
#XGBOOST - MODEL 2
from xgboost import XGBClassifier
#LIGHTGBM - MODEL 3
from lightgbm import LGBMClassifier

from sklearn.model_selection import RepeatedStratifiedKFold, cross_val_score
from scipy import stats


#permutation importance
from sklearn.inspection import permutation_importance, PartialDependenceDisplay

from sklearn.metrics import f1_score, recall_score, precision_score, roc_curve, roc_auc_score, average_precision_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# SHAP 
# Automatic SHAP explainer for final model
import shap
from sklearn.ensemble import RandomForestClassifier
from sklearn.base import is_classifier




df = pd.read_csv("/kaggle/input/datasets/organizations/uciml/pima-indians-diabetes-database/diabetes.csv")
X = df.drop("Outcome",axis = 1)
y = df["Outcome"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify = y,random_state = 42)
print(y_test.value_counts())
print(y_train.value_counts())

#target distribution
sns.countplot(x=y_train)
plt.show()

#feature distribution
X_train.hist(bins=20,figsize=(12,8))
plt.show()

#correlations heatmap
plt.figure(figsize=(12,8))
sns.heatmap(X_train.corr(),annot=True, cmap="coolwarm",center=0)
plt.show()

#feature distribution related to target values
for col in X_train.columns:
    sns.boxplot(x=y_train,y=X_train[col])
    plt.show()

#check for values equal to 0
(X_train==0).sum()

#replace 0 with NaN
missing_features =['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
X_train[missing_features]=X_train[missing_features].replace(0,np.nan)
X_test[missing_features]=X_test[missing_features].replace(0,np.nan)

#recheck for safety
X_train.isna().sum()



#LOGISTIC REGRESSOR - BASELINE
pipeline_lr = Pipeline([('imputer',IterativeImputer(max_iter=10,random_state=42)),
                       ('scaler', StandardScaler()),
                       ('model',LogisticRegression(max_iter=1000,random_state=42))])

param_grid_lr = {
                'model__C': [0.01,0.1,1,10],
                'model__penalty':['l2'],
                'model__solver':['lbfgs'],
                'model__class_weight':[None,'balanced']}

grid_lr = GridSearchCV(estimator=pipeline_lr,
                      param_grid=param_grid_lr,
                      cv=5,
                      scoring={
                          'f1':'f1',
                          'accuracy':'accuracy',
                          'recall':'recall'
                      },
                      refit = 'f1',
                      n_jobs=-1)

grid_lr.fit(X_train, y_train)

print("Best F1 - Logistic: ", grid_lr.best_score_)
print("Std dev - Logistic: ",grid_lr.cv_results_['std_test_f1'][grid_lr.best_index_])
print("Best params - Logistic: ", grid_lr.best_params_)
best_lr = grid_lr.best_estimator_
print("Recall - Logistic: ",grid_lr.cv_results_['mean_test_recall'][grid_lr.best_index_])
print("Accuracy - Logistic: ",grid_lr.cv_results_['mean_test_accuracy'][grid_lr.best_index_])


#MLP PIPELINE
pipeline_mlp = Pipeline([
    ('imputer',IterativeImputer(max_iter=10,random_state=42)),
    ('scaler',StandardScaler()),
    ('model',MLPClassifier(
        max_iter=1000,
        early_stopping=True,
        random_state=42
    ))
])

param_grid_mlp = {
    'model__hidden_layer_sizes':[(100,),(200,),(200,150)],
    'model__alpha': [0.001,0.003,0.005],
    'model__activation': ['relu','tanh'],
    'model__learning_rate_init':[0.001,0.002,0.003]
}

grid_mlp = GridSearchCV(
                      estimator=pipeline_mlp,
                      param_grid=param_grid_mlp,
                      cv=5,
                      scoring={
                          'f1':'f1',
                          'accuracy':'accuracy',
                          'recall':'recall'
                      },
                      refit = 'f1',
                      n_jobs=-1,
                      )

print(X_train.shape)
grid_mlp.fit(X_train, y_train)
print("Best F1 - MLP: ", grid_mlp.best_score_)
print("Std dev - MLP: ",grid_mlp.cv_results_['std_test_f1'][grid_mlp.best_index_])
print("Best params - MLP: ", grid_mlp.best_params_)
best_mlp = grid_mlp.best_estimator_
print("Recall - MLP: ",grid_mlp.cv_results_['mean_test_recall'][grid_mlp.best_index_])
print("Accuracy - MLP: ",grid_mlp.cv_results_['mean_test_accuracy'][grid_mlp.best_index_])






ratio = (y_train == 0).sum() / (y_train == 1).sum()

#XGBOOST PIPELINE
pipeline_xgb = Pipeline([
    ('imputer',IterativeImputer(max_iter=10,random_state=42)),
    (
    'model',XGBClassifier(
        random_state = 42,
        eval_metric = 'logloss')
    )
])

param_grid_xgb = {
    'model__n_estimators': [100,200],
    'model__max_depth':[3,4,5],
    'model__learning_rate':[0.05,0.1],
    'model__subsample':[0.8,1.0],
    'model__colsample_bytree':[0.8,1.0],
    'model__scale_pos_weight':[1,ratio]
}

grid_xgb = GridSearchCV(
    estimator = pipeline_xgb,
    param_grid = param_grid_xgb,
    scoring = {
        'f1':'f1',
        'accuracy':'accuracy',
        'recall':'recall'
    },
    refit = 'f1',
    cv = 5,
    n_jobs = -1
)


grid_xgb.fit(X_train,y_train)

print("Best F1 - XGBoost: ",grid_xgb.best_score_)
print("Std dev - XGBoost: ",grid_xgb.cv_results_['std_test_f1'][grid_xgb.best_index_])
print("Best params - XGBoost: ",grid_xgb.best_params_)
best_xgb = grid_xgb.best_estimator_
print("Recall - XGBoost: ",grid_xgb.cv_results_['mean_test_recall'][grid_xgb.best_index_])
print("Accuracy - XGBoost: ",grid_xgb.cv_results_['mean_test_accuracy'][grid_xgb.best_index_])



#LIGHTGBM PIPELINE
pipeline_lgbm = Pipeline ([(
    'imputer',IterativeImputer(max_iter=10,random_state = 42)),
    (
        'model',LGBMClassifier(random_state=42,verbose=-1,
                               )
    )
])

param_grid_lgbm = {
    'model__num_leaves': [31,50,100],
    'model__max_depth': [-1,5,10],
    'model__learning_rate': [0.01,0.05,0.1],
    'model__n_estimators': [100,200,500],
    'model__class_weight': [None, 'balanced']
}

grid_lgbm = GridSearchCV(
    estimator = pipeline_lgbm,
    param_grid = param_grid_lgbm,
    scoring = {
        'f1':'f1',
        'accuracy':'accuracy',
        'recall':'recall'
    },
    refit = 'f1',
    cv = 5,
    n_jobs = -1,
    verbose = 0
)

grid_lgbm.fit(X_train, y_train)
print("Best F1 - LightGBM: ",grid_lgbm.best_score_)
print("Std dev - LightGBM: ",grid_lgbm.cv_results_['std_test_f1'][grid_lgbm.best_index_])
print("Best params - LightGBM: ",grid_lgbm.best_params_)
best_lgbm = grid_lgbm.best_estimator_
print("Recall - LightGBM: ",grid_lgbm.cv_results_['mean_test_recall'][grid_lgbm.best_index_])
print("Accuracy - LightGBM: ",grid_lgbm.cv_results_['mean_test_accuracy'][grid_lgbm.best_index_])


#  Statistical analysis (Corrected Resampled t-test)

print("\nStart of advanced statistical validation (10-fold CV repeated 10 times)...")

# 1. Configuration of Repeated Cross-Validation  (10 fold x 10 times = 100 statistical samples)
n_splits = 10
n_repeats = 10
cv_ripetuta = RepeatedStratifiedKFold(n_splits=n_splits, n_repeats=n_repeats, random_state=42)

# Dictionary with the best estimators extracted from GridSearchCV pipelines 
mappa_modelli = {
    "MLP": grid_mlp.best_estimator_,
    "LGBM": grid_lgbm.best_estimator_,
    "XGB": grid_xgb.best_estimator_
}

# 2. Extraction of the 100 F1-scores for each of the three models

print("Computation of the 100 folds for MLP (Neural Network)...")
scores_mlp = cross_val_score(mappa_modelli["MLP"], X_train, y_train, cv=cv_ripetuta, scoring='f1', n_jobs=-1)

print("Computation of the 100 folds for LightGBM...")
scores_lgbm = cross_val_score(mappa_modelli["LGBM"], X_train, y_train, cv=cv_ripetuta, scoring='f1', n_jobs=-1)

print("Computation of the 100 folds for XGBoost...")
scores_xgb = cross_val_score(mappa_modelli["XGB"], X_train, y_train, cv=cv_ripetuta, scoring='f1', n_jobs=-1)

# Average values on the 100 folds for the three models
punteggi_100_fold = {
    "MLP": np.mean(scores_mlp),
    "LGBM": np.mean(scores_lgbm),
    "XGB": np.mean(scores_xgb)
}

plt.figure(figsize=(8, 5))
# Boxplot representation to visualize F1-scores distributions
sns.boxplot(data=[scores_mlp, scores_lgbm, scores_xgb], palette="Set2")
plt.xticks([0, 1, 2], ['MLP (Rete Neurale)', 'LightGBM', 'XGBoost'])
plt.ylabel('F1-Score on the 100 folds')
plt.title('Distribution of Performance in the 100 Validation Folds')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Ordering of the models by scores
classifica = sorted(punteggi_100_fold.items(), key=lambda x: x[1], reverse=True)
primo_nome, primo_score = classifica[0]
secondo_nome, secondo_score = classifica[1]
terzo_nome, third_score = classifica[2]

print(f"\nReal classification based on the 100 folds (average F1-score):")
print(f"1° place: {primo_nome} ({primo_score:.4f})")
print(f"2° place: {secondo_nome} ({secondo_score:.4f})")
print(f"3° place: {terzo_nome} ({third_score:.4f})")

# 3. Internal function to compute Corrected t-test between two score vectors
def calcola_corrected_p_value(scores_A, scores_B, n_splits=10):
    k = len(scores_A)                                 # N. of samples (100)
    differences = scores_A - scores_B                 # Differences d_i
    m_d = np.mean(differences)                        # Average of differences
    var_d = np.var(differences, ddof=1)               # Variance of differences sigma_d^2
    
    # test/train set ratio (0.1 / 0.9 for 10-fold CV)
    n2_su_n1 = (1 / n_splits) / (1 - (1 / n_splits)) 
    
    # Correct statistical formula for related samples
    t_stat = m_d / np.sqrt(((1 / k) + n2_su_n1) * var_d)
    
    # Two-tailed p-value with k-1 degrees of freedom
    p_val = 2 * (1 - stats.t.cdf(np.abs(t_stat), k - 1))
    return p_val

# 4. Statistical tests: 1° classified model vs. 2° and 3°
p_val_vs_2 = calcola_corrected_p_value(
    scores_mlp if primo_nome=="MLP" else (scores_lgbm if primo_nome=="LGBM" else scores_xgb),
    scores_mlp if secondo_nome=="MLP" else (scores_lgbm if secondo_nome=="LGBM" else scores_xgb)
)

p_val_vs_3 = calcola_corrected_p_value(
    scores_mlp if primo_nome=="MLP" else (scores_lgbm if primo_nome=="LGBM" else scores_xgb),
    scores_mlp if terzo_nome=="MLP" else (scores_lgbm if terzo_nome=="LGBM" else scores_xgb)
)

print("\n" + "="*60)
print("       VERDICT OF SIGNIFICANCE TESTS")
print("="*60)
print(f"Test {primo_nome} vs {secondo_nome} -> P-Value: {p_val_vs_2:.6f}")
print(f"Test {primo_nome} vs {terzo_nome}   -> P-Value: {p_val_vs_3:.6f}")
print("-"*60)

# 5. Final decisional logic (combines statistical significance with clinical criterions)
# If the first model has a p-value < 0.05 vs. the second one, it wins in a statistically significant way
if p_val_vs_2 < 0.05:
    print(f"RESULT: {primo_nome} is statistically better than the second model.")
    winner_name = primo_nome
    best_model = mappa_modelli[primo_nome]
else:
    # if p-value >= 0.05, there's a statistical tie. Tie-breaker based on Recall is employed
    print(f"RESULT: Statistical tie between {primo_nome} and {secondo_nome}.")
    print("Tie-breaker: medical criterion (Recall Maximization):")
    
    # Recovery of the recall scores from the pipelines to solve the tie 
    recalls = {
        "MLP": grid_mlp.cv_results_['mean_test_recall'][grid_mlp.best_index_],
        "LGBM": grid_lgbm.cv_results_['mean_test_recall'][grid_lgbm.best_index_],
        "XGB": grid_xgb.cv_results_['mean_test_recall'][grid_xgb.best_index_]
    }
    
    print(f"-> Recall {primo_nome}:   {recalls[primo_nome]:.4f}")
    print(f"-> Recall {secondo_nome}: {recalls[secondo_nome]:.4f}")
    
    if recalls[primo_nome] >= recalls[secondo_nome]:
        winner_name = primo_nome
        best_model = mappa_modelli[primo_nome]
    else:
        winner_name = secondo_nome
        best_model = mappa_modelli[secondo_nome]
        
    print(f"\nReal final winner (clinical choice): {winner_name}")
print("="*60 + "\n")



#refit on whole training set
best_model.fit(X_train, y_train)

#test with the chosen threshold
y_test_pred = best_model.predict(X_test)
y_probs = best_model.predict_proba(X_test)[:, 1]

#results
print("Test F1:", f1_score(y_test, y_test_pred))
print("Test Recall:", recall_score(y_test, y_test_pred))
print("Test Precision: ", precision_score(y_test, y_test_pred))
print("Average Precision:", average_precision_score(y_test,y_test_pred))



cm = confusion_matrix(y_test, y_test_pred, labels=[0, 1])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Healthy (0)', 'Diabetic (1)'])
disp.plot(cmap='Blues')



# Computation of AUC
auc_score = roc_auc_score(y_test, y_probs)

# Computation of ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# Plot
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, linewidth=2, label=f'ROC Curve (AUC = {auc_score:.3f})')
plt.plot([0,1], [0,1], linestyle='--', alpha=0.6, label='Random classifier')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"ROC AUC: {auc_score:.3f}")







# Computation of permutation importance
perm_result = permutation_importance(
    best_model, X_test, y_test,
    n_repeats=10,  
    random_state=42,
    scoring='f1'   
)

# Prepare DataFrame ordered by importance
features = X_test.columns if hasattr(X_test, "columns") else [f"feature_{i}" for i in range(X_test.shape[1])]
perm_df = pd.DataFrame({
    'feature': features,
    'importance': perm_result.importances_mean,
    'importance_std_dev' : perm_result.importances_std
})
perm_df = perm_df.sort_values(by='importance', ascending=False)

# Print top 10 feature
print(perm_df.head(10))

# Plot top 10
perm_df.head(10).plot.bar(x='feature', y='importance', legend=False, figsize=(10,5))
plt.title(f"Permutation Feature Importance (F1) - {winner_name}")
plt.ylabel("Mean decrease in F1")
plt.show()




# 1. Select top features (from perm_df)
features_to_plot = perm_df['feature'].iloc[:8].tolist()

# 2. Create plot grid
fig, ax = plt.subplots(figsize=(18, 12))

# 3. Generate PDP (yellow line) and ICE (multiple thin blue lines)
display = PartialDependenceDisplay.from_estimator(
    best_model, 
    X_test, 
    features=features_to_plot,
    kind='both', # displays PDP (average) and ICE (individual) lines
    subsample=50, # N. of ICE lines for chart
    n_cols=3,
    ax=ax,
    pd_line_kw={'color': 'gold', 'linewidth': 4},
    ice_lines_kw={'color': 'skyblue', 'alpha': 0.3}
)

# 4. BICOLOR RUGPLOT 
for i, feature_name in enumerate(features_to_plot):
    current_ax = display.axes_.flatten()[i]
    
    # Healthy (Class 0) - BluE
    sns.rugplot(
        X_test[y_test == 0][feature_name], 
        color='blue', alpha=0.3, height=0.06, ax=current_ax, label='Healthy'
    )
    
    # Diabetic (Class 1) - Red
    sns.rugplot(
        X_test[y_test == 1][feature_name], 
        color='red', alpha=0.5, height=0.09, ax=current_ax, label='Diabetic'
    )
    

    
    if i == 0:
        current_ax.legend(loc='upper left', fontsize='small')
  
    

scaler_info = "+ Scaler" if "MLP" in winner_name else ""
# Create a suffix to know if the data has been scaled (useful for MLP)
plt.suptitle(f"{winner_name} Analysis: PDP + ICE + Rugplot {scaler_info} (Healthy=Blue, Diabetic=Red)", 
             fontsize=20, y=1.02)
plt.tight_layout()
plt.show()




# Extract preprocessing and model from the pipeline
preprocessor = best_model.named_steps['imputer']
model = best_model.named_steps['model']

# Trasforms test data
X_test_transformed = preprocessor.transform(X_test)





feature_names = X_test.columns if hasattr(X_test, "columns") else [f"feature_{i}" for i in range(X_test.shape[1])]

# --- Choose the explainer automatically ---
if isinstance(model, (LGBMClassifier, XGBClassifier, RandomForestClassifier)):
    explainer = shap.TreeExplainer(model)
    
    ev = explainer.expected_value
    
    # Robust handling to extract scalar value
    if isinstance(ev, (list, np.ndarray)):
        if len(ev) > 1:
            # Takes the element 1 and converts it into a simple number
            valore_base = ev[1].item() if hasattr(ev[1], 'item') else ev[1]
        else:
                # Takes the only element
            valore_base = ev[0].item() if hasattr(ev[0], 'item') else ev[0]
    else:
        valore_base = ev

    
    print(f"Base value (average starting point): {valore_base:.2f}")
    
    # 2. Calcolo shap_values
    shap_values = explainer.shap_values(X_test_transformed)
else:
    scaler=best_model.named_steps['scaler']
    X_test_transformed = scaler.transform(X_test_transformed)
    # model-agnostic for MLP, Logistic, SVM
    def model_predict(X):
        return model.predict_proba(X)[:,1]
    background = X_test_transformed[np.random.choice(X_test_transformed.shape[0], 100, replace=False)]
    explainer = shap.KernelExplainer(model_predict, background)
    shap_values = explainer.shap_values(X_test_transformed[:50])  # subset per velocità

# --- Global Summary plot ---
shap.summary_plot(shap_values, X_test_transformed if not isinstance(explainer, shap.KernelExplainer) else X_test_transformed[:50],
                  feature_names=feature_names, plot_type="bar")  # absolute value scores

shap.summary_plot(shap_values, X_test_transformed if not isinstance(explainer, shap.KernelExplainer) else X_test_transformed[:50],
                  feature_names=feature_names)  # more detailed

# 1. Patient index chosen randomly and then confirmed if it is considered as a statistically interesting sample to explain

paziente_idx = 12


exp = shap.Explanation(
    values=shap_values[paziente_idx], 
    base_values=explainer.expected_value[1] if isinstance(shap_values, list) else explainer.expected_value,
    data=X_test_transformed[paziente_idx],
    feature_names=feature_names
)

# 3. Plot Waterfall 
shap.plots.waterfall(exp, show=False)
fig = plt.gcf() 
fig.set_size_inches(12, 6) 
plt.title(f"Local interpretation: Patient {paziente_idx} ({winner_name})", pad=30, fontsize=15)
plt.show()


```
